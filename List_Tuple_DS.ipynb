# List in data structures
1. list is collection of elements enclosed by []
2. list a mutable - editable
3. it's allowed duplicate elements
4. each element seperated by,

```python
# create list
names=["shiva","vikram","vikas","divya"]
ages=[25,27,28.9,27.9]
print(names)
print(ages)
print(type(names))
print(type(ages))
      
```

```python
#concatenation of list
a=[10,20,30,40,50]
b=[40,50,60,70,80]
result=a+b
print(result)
```

```python
#repeating list
ages=[10,20,30,40,50]
print(ages*10)
```

```python
#accesing elements from list - index
ages=[20,23,64,34,19,11,53]
print(ages)
print(ages[4])
     
```

```python
#methods or functions in list
print(dir(list))
```

```python
#description
print(help(list))
```

```python
#add elements to list = append(),extend(),insert()
a=[10,20,30,40,50]
print(a)
a.append(200)
print(a)
```

```python
a=[10,20,20,30]
print(a)
a.extend([100,200,300])
print(a)
```

```python
a=[10,20,20,30]
print(a)
a.insert(1,100)
print(a)
```

```python
#remove elements from list - pop(),remove(),clear()
a=[10,20,30,40,50]
print(a)
a.pop(4)
print(a)
```

```python
a=[10,20,30,40,50]
print(a)
a.remove(50)
print(a)
```

```python
#sorting techniques - sort(reverse=true or reverse=false)
a=[6,5,4,3,1,2]
print(a)
a.sort(reverse=False) # True means descending , false means ascending
print(a)
```

# Tuples in data structures
1. tuple is a collection of elements enclosed by ()
2. tuple is immutable - not editable
3. each element seperated
4. duplicates allowed

```python
a=(10,20,30,40,50)
names=("shiva","vikram",10.25,12.56)
print(a)
print(names)
print(type(a))
print(type(names))
```

```python
#concatenation,repeating,Accessing elements from tuple
```

```python
#concatenation of tuple

t1 = (1, 2, 3)
t2 = (4, 5, 6)
result = t1 + t2
print(result)
```

```python
#repeating of tuple

t = (1, 2, 3)
result = t * 3
print(result)


```

```python
#Accessing elements from tuple - index

t=(1,2,3,4,5,6,9,8)
print(t)
print(t[4])
print(t[-3])

```

```python
#methods 
print(dir(tuple))
```

```python
#description 
print(help(tuple))
```

```python
# count()
a=(10,10,20,30,40,50,10)
print(a.count(10))
```

```python
# index()
a=(10,20,30,40,50)
print(a.index(40))
```
