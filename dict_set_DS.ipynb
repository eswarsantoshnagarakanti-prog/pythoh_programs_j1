# dict
1. dict is collection of items enclosed by {} - item is a key:value pair

```python
info={"name":"shiva","salary":1000.25,"address":"hyderabad","age":25}
print(info)
print(type(info))
```

```python
# read data and update data from dict
info={"name":"siva","salary":10000.25,"address":"hyderabad","age":25}
print(info)
print(info["address"])
```

```python
info={"name":"siva","salary":10000.25,"address":"hyderabad","age":25}
print(info)
info["address"]="banglore"
print(info)
```

```python
#methods or function in dict
print(dir(dict))
```

```python
#description
print(help(dict))
```

```python
info={"name":"siva","salary":10000.25,"address":"hyderabad","age":25}
print(info)
print(info.keys())
print(info.values())
print(info.items())
```

```python
#update()
info={"name":"siva","salary":10000.25,"address":"hyderabad","age":25}
print(info)
info.update({"salary":10000000.50})
print(info)
```

```python
#pop()
info={"name":"siva","salary":10000.25,"address":"hyderabad","age":25}
print(info)
info.pop("salary")
print(info)
```

# sets
1. set is a collection of unorder unique elements enclosed by {}

```python
# create set
a={2,4,1,3,5,6,7,1,2,3}
print(a)
print(type(a))
```

```python
# methods or functions
print(dir(set))
```

```python
print(help(set))
```

```python
# add()
a={20,30,40,50}
print(a)
a.add(100)
print(a)
```

```python
#difference()
a={1,2,3,4,5}
b={23,45,43,32,43}
c=b.difference(a)
print(c)
```

```python
a={1,2,3,4,5,6}
b={4,5,6,7,8,9}
c=a.union(b)
print(c)
```

```python
#intersection
a={1,2,3,4,5,6}
b={4,5,6,7,8,9}
c=a.intersection(b)
print(c)
```

```python
#Discard
a={1,2,3,4,5,6}
print(a)
a.discard(6)
print(a)
```
