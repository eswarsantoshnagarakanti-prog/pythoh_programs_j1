````python
""" write a program to find the grades based on the marks
1.if marks in between 80-100 - grade is A
2.if marks are in between 60-79 - grade is B
3.if marks are in between 40-59 - grade is C
4.if marks are in between 25-39 - grade is D
5.if marks in between 0-24 - grade is fail
6.if marks other than above output then - please enter correct marks
"""
marks=int(input("Please enter your marks"))
if(marks>=80 and marks<=100):
    print("Grade is A")
elif(marks>=60 and marks<=79):
    print("Grade is B")
elif(marks>=40 and marks<=59):
    print("Grade is C")
elif(marks>=25 and marks<=39):
    print("Grade is D")
elif(marks>=0 and marks<=24):
    print("Grade is fail")
else:
    print("print enter the correct marks")
    
````
